```python
import requests

import pandas as pd

from bs4 import BeautifulSoup
```

```python
def get_items(array):
    items = []
    for element in array:
        item = element.text
        try:
            item = float(item)
        except:
            pass
        items.append(item)
    return items
```

```python
page = requests.get("https://opencatalystproject.org/leaderboard_is2re.html").text
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
table = soup.find(id="leaderboard")
```

```python
head_flag = True
table_lists = []
for row in table.find_all('tr'):
    tds = row.find_all('td')
    ths = row.find_all('th')
    row_items = get_items(tds + ths)
    table_lists.append(row_items)
```

```python
leaderboard = pd.DataFrame(table_lists[1:], columns=table_lists[0][:2] + [table_lists[0][-1]] + table_lists[0][2:-1])
```

```python
leaderboard['MAE'] = (leaderboard['Energy MAE (eV) -- ID']*24943
                      + leaderboard['Energy MAE (eV) -- OOD Ads']*24961
                      + leaderboard['Energy MAE (eV) -- OOD Cat']*24963
                      + leaderboard['Energy MAE (eV) -- OOD Both']*24987) / (24943+24961+24963+24987)
```

```python
leaderboard = leaderboard.sort_values(by='MAE')
```
